```python
import nltk, spacy
nltk.download("punkt")
spacy_en = spacy.load("en_core_web_sm")

from nltk.tokenize import word_tokenize
```

### Tokenization

```python
en_text = "This is a sample sentence for tokenization test"

print(f"Tokenized by nltk : {word_tokenize(en_text)}")
print(f"Tokenized by spacy : {[token.text for token in spacy_en.tokenizer(en_text)]}")
```
